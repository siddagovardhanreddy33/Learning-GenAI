```python
# client
client = genai.Client() # genai.client(api-key = 'key')
client
```

```python
# Interact Gemini LLM - gemini-2.5-flash-lite

response = client.models.generate_content(
  model='gemini-2.5-flash-lite',
  contents='what are the advantages of Gemini 3 ovar Gemini 2.5',  #prompt[USER/HUMAN]
)

print(response.text)
```

```python
# Interact Gemini LLM - gemini-2.5-flash-lite


# We can customize the generation through Prompts
response = client.models.generate_content(
  model='gemini-2.5-flash-lite',
  contents='what are the 3 advantages of Gemini 3 ovar Gemini 2.5 in 3 sentences ',  #prompt[USER/HUMAN]
)

print(response.text)
```

```python
# Interact Gemini LLM - gemini-2.5-flash-lite

# FEB 15 2024 is the cut of date for gemini 2.5 , it can answer only before feb 15 2024

response = client.models.generate_content(
  model='gemini-2.5-flash-lite',
  contents='when is gemini 2.5 updated recently ',  #prompt[USER/HUMAN]
)

print(response.text)

# if we need recent info , then we have to add few tools to it
```

```python
# Interact Gemini LLM - gemini-2.5-flash-lite

response = client.models.generate_content(
  model='gemini-2.5-flash-lite',
  contents='when is gemini 2.5 updated recently , respond in  single sentance',  #prompt[USER/HUMAN]
)

print(response.text)
```

### System Prompt
- we can assigne a role to LLM , it decides the behaviour of LLM

```python
## to fix the behaviour of an LLM to respond in 2 to 3 sentences : system prompt


response = client.models.generate_content(
    model='gemini-2.5-flash-lite',
    contents='When is Gemini 2.5 updated recently?',
    config=genai.types.GenerateContentConfig(
        system_instruction='always respond to the user queries concisely ', # System prompt or system message - fixes the behaviour of llm
    )
)
print(response.text)
```

## Quick Tips for Getting Started

1. **Identify Repetitive Content Tasks** – Look for places where humans spend a lot of time creating similar text, images, or code. Those are low‑hanging fruit for automation.  
2. **Start Small with a Pilot** – Use a pre‑trained model (e.g., ChatGPT, Stable Diffusion) to prototype a single workflow before building a custom solution.  
3. **Measure Impact** – Track metrics like time saved, cost reduction, quality scores, and user satisfaction to justify further investment.  
4. **Blend Human & AI Strengths** – Treat generative AI as an “assistant” that drafts and augments, while humans perform the final curation and critical decision‑making.  
5. **Stay Ethical & Compliant** – Ensure generated content respects copyright, privacy, and bias guidelines; maintain transparency with end‑users when AI is involved.

```python
# Openai - interaction with GPT models

response = client.chat.completions.create(
    messages = [            # In point of groq prompt is a message that list of [user,system]
        {
            'role': 'user',
            'content': 'what are the advantagess of Generative AI',
        },
        {
            'role':'system',
            'content': 'always respond to the user queries concisely'
        }

    ],
    model = 'openai/gpt-oss-120b'  #model
)

response
```
