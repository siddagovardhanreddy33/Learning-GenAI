## Interacting with gemini LLM

```python
# create a

!pip install -U langchain-google-genai
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
import os
from google.colab import userdata
```

```python
gem = userdata.get('gemini_genAI')
os.environ['GEMINI_API_KEY'] = gem
```

```python
model = ChatGoogleGenerativeAI(
    model = 'gemini-2.5-flash-lite'
)

model
```

```python
messages = [
    ("system","You are a helpful assistant that translates English to Hindi. Translate the user sentence.",),
    ("human", "I love programming."),
]


messages
```

```python
model.invoke(messages)
```

```python
responce = model.invoke(messages)
responce
```

```python
responce.content
```
